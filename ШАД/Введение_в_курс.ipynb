Реализации:
- Cpython
- IronPython
- Juthon
- PyPy

Базовые типы:
- Числовые типы (int, float, complex)
- Переменные (динамическая типизация, называй не однобуквенно)
- Логический тип (bool)
    - Тrue, False (not, and, or)
- Строковый тип (str)
    - тройные кавычки - сохраняют структуру текста (абзацы и тп символы переноса)
    - строки, срезы, подстроки (можно обращаться к отдельным элементам)
    - правый конец интервала не включается - [a, b)
- Словарь (dict):
    - во всех щелях, имба
    - ключ - значения
- Списки (list):
    - коллекция ссылок на объект, коллекция объектов, с которыми нужно одинаково работать
- Кортеж (tuple):
    - похож на список, но смысл совершенно иной
    - коллекция объектов, разных типов, но одинакового размера
    - неизменяемый 

переменная _ записывает результат последнего выражения
