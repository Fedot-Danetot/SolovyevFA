4.3.2 - Дана функция большая-большая; Создать массив из 10 значений функции. Выделить срез первой половины массива и построить графики для основного массива, используя plt.plot(), для среза - точечный (plt.scatter())

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 10, 1)
y = np.abs(np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(np.sqrt(x)))/np.fabs(1 - np.power(np.sin(x), 3)) + np.log(np.fabs(2 * x)))
plt.plot(x, y)

plt.scatter(x[:6], y[:6])
```

4.3.3 - Дана функция поменбше. Построить график на интервале (0, 10) с шагом 1 с заливкой площади и найти эту площадь под ним. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simpson
from numpy import trapz

x = np.arange(0.0, 10, 1)
y = np.abs(np.fabs(np.cos(x*np.exp(np.cos(x) + np.log(x + 1)))))

plt.plot(x, y, c = "navy")
plt.fill_between(x, y)

area = trapz(y)
print(area)
```

4.3.4 - Построить 3 графика на плоскости о стоимости акций трех компаний

```python
import numpy as np
import matplotlib.pyplot as plt

x = ['январь','февраль','март','апрель','май','июнь','июль','август','сентябрь','октябрь','ноябрь','декабрь']
y1 = [131.96, 121.26, 122.15, 131.46, 124.61, 136.96, 145.86, 151.83, 141.50, 149.80, 165.30, 177.57]
y2 = [231.96, 232.38, 235.77, 252.18, 249.68, 270.90, 284.91, 301.88, 281.92, 331.62, 330.59, 336.32]
y3 = [91.79, 101.84, 103.43, 120.51, 120.58, 125.32, 135.22, 145.46, 133.27, 148.27, 142.45, 144.68]

plt.plot(x, y1, c = "r")
plt.plot(x, y2, c = "g")
plt.plot(x, y3, c = "b")

plt.xlabel('месяц')
plt.ylabel('стоимость акций, доллары')
plt.title('Акции крупных компаний за 2021 год')

plt.bar(x, y1, color = "r", label = 'Apple', alpha = 0.5)
plt.bar(x, y2, color = "g", label = 'Microsoft', alpha = 0.5)
plt.bar(x, y3, color = "b", label = 'Google', alpha = 0.5)

plt.xticks([x+0.35 for x in range(12)], rotation='vertical')

plt.show()
```
